```python
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_pil_image
from torchvision.datasets import ImageFolder
from torchvision.transforms import Resize, ToTensor, Compose

transform = Compose([
    Resize((224, 224)),
    ToTensor()
])

classes = [
    'melanoma',
    'nevus',
    'basal_cell_carcinoma',
    'actinic_keratosis',
    'benign_keratosis',
    'dermatofibroma',
    'vascular_lesion',
    'squamous_cell_carcinoma'
]

train_root = 'BCN_20k_train'
train_Dataset = ImageFolder(root='BCN_20k_train/BCN_20k_train', transform=transform)
train_loader = DataLoader(train_Dataset, batch_size=32, shuffle=True)

train_class_names = train_Dataset.classes
print("TRAIN")
print(f"Classes_train: {train_class_names}")
print(f"Train samples: {len(train_Dataset)}")

test_root = 'BCN_20k_test'
test_Dataset = ImageFolder(root='BCN_20k_test/_20k_test', transform=transform)
test_loader = DataLoader(test_Dataset, batch_size=32, shuffle=False)

test_class_names = test_Dataset.classes
print("\nTEST")
print(f"Classes_test: {test_class_names}")
print(f"Test samples: {len(test_Dataset)}")

# Visualization function
def show_batch(loader, class_names, title="Sample Images", n=4):
    images, labels = next(iter(loader))

    fig, axes = plt.subplots(1, n, figsize=(12, 3))
    fig.suptitle(title)

    for i in range(n):
        axes[i].imshow(to_pil_image(images[i]))
        axes[i].set_title(class_names[labels[i].item()])
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

# Show sample images from the training set
show_batch(train_loader, train_class_names, title="Train Classes Sample Images")
# Show sample images from the test set
show_batch(test_loader, test_class_names, title="Test Classes Sample Images")

print(train_Dataset.classes)
print(test_Dataset.classes)
```
